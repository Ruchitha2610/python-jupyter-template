### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)  # X values from 0 to 10
y = x ** 2                   # Y values as the square of X

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = x^2')
plt.title('Line Plot of y = x^2')
plt.xlabel('X values')
plt.ylabel('Y values (Square of X)')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
